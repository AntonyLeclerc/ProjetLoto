```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.signal import convolve2d
```

```python
sharpen_kernel = np.array([1,4,6,4,1,4,16,24,16,4,6,4,-476,24,6,4,16,24,16,4,1,4,6,4,1]).reshape((5,5))
```

```python
def openImage(f):
    return np.array(Image.open(f).convert(mode="L"))
```

```python
flou13_1 = openImage("TargetSet/num28/img_1634731447_0013_0.jpg")
flou13_2 = openImage("TargetSet/num28/img_1634731447_0013_1.jpg")
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(flou13_1, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(flou13_2, cmap="gray")
```

```python
tmp1 = convolve2d(flou13_1, sharpen_kernel, mode="same")
tmp2 = convolve2d(flou13_2, sharpen_kernel, mode="same")
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.imshow(flou13_1, cmap="gray")
plt.subplot(1,2,2)
plt.imshow(tmp1, cmap="gray")
```

```python
plt.imshow(np.where(tmp1 < 150, 0, 255), cmap="gray")
```
